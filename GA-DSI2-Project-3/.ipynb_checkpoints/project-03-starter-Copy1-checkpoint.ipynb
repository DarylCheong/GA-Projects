```python
# split train/test sets

from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)
print X_train.shape
print y_train.shape
print y_train.value_counts()
print X_test.shape
print y_test.shape
print y_test.value_counts()
```

# Model 1 - Logistic Regression

```python
# function to create confusion matrix

def print_cm(y_true, y_pred):
    confusion = pd.crosstab(y_true, y_pred, rownames=['Actual'], colnames=['Predicted'], margins=True)
    print confusion
```

```python
# use logistic regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

logreg1 = LogisticRegression()

logreg1.fit(X_train, y_train)
print 'training accuracy: ' + str(logreg1.score(X_train, y_train))

y1_pred = logreg1.predict(X_test)
print 'testing accuracy: ' + str(logreg1.score(X_test, y_test))

print classification_report(y_test, y1_pred)
print_cm(y_test, y1_pred)
```

```python
# calculate ROC score

y1_proba = logreg1.predict_proba(X_test)
y1_proba = [p[1] for p in y1_proba]
print roc_auc_score(y_test, y1_proba)
```
