# DSI2 Project 4 - Salary Model Building 

```python
import pandas as pd
import numpy as np
import pickle

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
%matplotlib inline

pd.set_option('display.max_columns', None)
```

```python
# Load salary pickle data.

pickle_in = open("salary.pickle","rb")
salary = pickle.load(pickle_in)
print salary.shape
salary.head()
```

# Prepare target y and predictor X

```python
# Plot the salary distribution.

ax = salary['salary'].plot(kind='hist')
ax.axvline(salary["salary"].median(), lw=2.5, ls='solid', color='black')
plt.title('Salary Distribution')
```

```python
# Create classification group based on median salary amount.

median_salary = np.median(salary['salary'])
print median_salary
```

```python
# Create new column for salary grade classification.

salary['salary_grade'] = salary.apply(lambda x: 1 if x['salary'] > median_salary else 0, axis=1)

salary.head()
```

```python
salary['salary_grade'].value_counts().plot(kind='bar')
```

```python
# Set target y 'salary_grade' and calculate baseline.

y = salary['salary_grade']

baseline = 1 - np.mean(y)
baseline
```

```python
# Set predictor X 'title' and perform CountVectorizer

from sklearn.feature_extraction.text import CountVectorizer

X = salary['title']

cvec = CountVectorizer(stop_words='english', ngram_range=(1, 2))

cvec.fit(X)
```

```python
# Create new dataframe with vectorized words.

X_trans = pd.DataFrame(cvec.transform(X).todense(), columns=cvec.get_feature_names())
X_trans.head()
```

```python
# View most common words.

word_counts = X_trans.sum(axis=0).sort_values(ascending = False)
word_counts.head(20)
```

```python
# Prepare train/test sets.

from sklearn.model_selection import train_test_split

print y.shape
print X_trans.shape

X_train, X_test, y_train, y_test = train_test_split(X_trans, y, test_size=0.3, random_state=10)

print y_train.shape
print X_train.shape
print y_test.shape
print X_test.shape
```

# Logistic Regression

```python
# Implement logistic regression.

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

lr = LogisticRegression()

lr.fit(X_train, y_train)
print lr.score(X_train, y_train)
print lr.score(X_test, y_test)
```

```python
# Function to create confusion matrix.

def print_cm(y_true, y_pred):
    confusion = pd.crosstab(y_true, y_pred, rownames=['Actual'], colnames=['Predicted'], margins=True)
    print confusion
```

```python
# View classification results.

y_pred = lr.predict(X_test)

print classification_report(y_test, y_pred)
print_cm(y_test, y_pred)
```

# Random Forest

```python
# Implement Random Forest classifier

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(200, random_state=59)

rfc.fit(X_train, y_train)
print rfc.score(X_train, y_train)
print rfc.score(X_test, y_test)
```

```python
# View classification results.

y_pred2 = rfc.predict(X_test)

print classification_report(y_test, y_pred2)
print_cm(y_test, y_pred2)
```

```python
feature_importance = pd.DataFrame(rfc.feature_importances_, index = X_trans.columns,
                                  ).reset_index()
feature_importance.columns = ['features', 'importance']
print feature_importance.shape
feature_importance.sort_values('importance', ascending=False).head(20)
```

# Predict jobs with no salary

```python
# Load jobs pickle data.

pickle_in = open("jobs.pickle","rb")
jobs = pickle.load(pickle_in)
```

```python
# Filter out jobs with no salary.

no_salary = jobs[jobs['salary'] == 'NA']
print no_salary.shape
no_salary.head()
```

```python
# Perform CountVectorizer on jobs with no salary.

X_pred = no_salary['title']

X_pred_trans = pd.DataFrame(cvec.transform(X_pred).todense(), columns=cvec.get_feature_names())

print X_pred_trans.shape
X_pred_trans.head()
```

```python
# Make predictions using the Random Forest model and append the results to the dataframe.

y_pred = rfc.predict(X_pred_trans)
X_pred_trans['salary_prediction'] = y_pred
X_pred_trans.head()
```
